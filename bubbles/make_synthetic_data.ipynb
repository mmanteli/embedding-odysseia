```python
import numpy as np
import json
```

```python


def generate_uniform_synthetic(X, num_samples=1000, random_state=None):


    rng = np.random.default_rng(random_state)
    mins = X.min(axis=0)
    maxs = X.max(axis=0)
    
    synthetic = rng.uniform(mins, maxs, size=(num_samples, X.shape[1]))
    return synthetic, mins, maxs


```

```python

X = []
with open("seed_embeddings.jsonl", "r") as f:
    lines = f.readlines()
    for l in lines:
        X.append(json.loads(l)["emb"])

X = np.array(X)
print(X.shape)

synthetic, mins, maxs = generate_uniform_synthetic(X, num_samples=1000, random_state=42)

print("mins:", mins)
print("maxs:", maxs)
print("Synthetic shape:", synthetic.shape)
```

```python
with open("seed_embeddings_synthetic.jsonl", "w") as f:
    for i,e in enumerate(synthetic):
        f.write(json.dumps({"id": "synth_"+str(i), "emb": e.tolist()})+"\n")
```
